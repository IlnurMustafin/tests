###4. Обучение модели.

**Для обучения будет использоваться FNN с одним входным слоем, с тремя скрытыми слоями (128, 64, 32 нейрона соответственно, функция активации - ReLu) и одним выходным слоем (функция активации - линейная).**

**Данные масштабируются при помощи StandardScaler и разбиваются на тренировочную и тестовую выборки.**

**Метрика для обучения - MAE.**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = result_cleaned.drop('quality', axis=1).values
y = result_cleaned['quality'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Масштабирование данных
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras import layers, models
from tensorflow.keras import metrics

# Создание модели
model = models.Sequential()
model.add(layers.Input(shape=(X_train.shape[1],)))  # Входной слой
model.add(layers.Dense(128, activation='relu'))              # Первый скрытый слой
model.add(layers.Dense(64, activation='relu'))               # Второй скрытый слой
model.add(layers.Dense(32, activation='relu'))               # Третий скрытый слой
model.add(layers.Dense(1, activation='linear'))              # Выходной слой


model.compile(optimizer='adam', loss='mean_squared_error', metrics=[metrics.MeanAbsoluteError()])
```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Оценка модели на тестовой выборке
mae_value = model.evaluate(X_test, y_test, verbose=0)  # verbose=0 чтобы не выводить дополнительные сообщения

# Первый элемент в возвращаемом списке - это значение потерь, второй - MAE
mae = mae_value[1]
print(f'Mean Absolute Error (MAE) на тестовой выборке: {mae}')
```

**Визуализируем качество модели.**

```python
# Визуализация графиков потери и метрик
plt.figure(figsize=(12, 5))

# График потерь
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# График MAE
plt.subplot(1, 2, 2)
plt.plot(history.history['mean_absolute_error'], label='Train MAE')
plt.plot(history.history['val_mean_absolute_error'], label='Validation MAE')
plt.title('MAE vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Mean Absolute Error')
plt.legend()

# Показать графики
plt.tight_layout()
plt.show()
```

```python
y_pred = model.predict(X_test)
```

```python
# Создание графика
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', label='Предсказанные значения')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--', label='Идеальное соответствие')
plt.xlabel('Истинные значения')
plt.ylabel('Предсказанные значения')
plt.title('Зависимость предсказанных оценок от истинных значений')
plt.legend()
plt.grid()
plt.axis('equal')
plt.show()
```
